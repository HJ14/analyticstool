## 3.7 Statistical transformations

ggplot2로 그리는 바차트 : 

geom_bar () 

데이터 : 가격, 캐럿, 색상, 선명도 및 각 다이아몬드의 컷을 포함하여 ~ 54,000 개의 다이아몬드에 대한 정보

차트를 통해 알 수 있는 것 :  품질이 낮은 절단보다 고품질의 절단으로 더 많은 다이아몬드를 사용할 수 있다. 

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut))
```

막대 그래프, 히스토그램 및 빈도 다각형을 사용하여 데이터를 저장 한 다음 각 빈에 속하는 빈 수를 플롯화.

모델에 데이터를 적용한 다음 예측할 수 있도록 플롯화.

박스 플롯은 분포를 요약한 다음 특수하게 포맷 된 박스를 표시해 줌.

<img src="2.png" width="900"><BR>

```r
ggplot(data = diamonds) + 
  stat_count(mapping = aes(x = cut))
```

 geom_bar ()의 통계를 count (기본값)에서 identity로 변경하면  
 
 막대의 높이를 yy 변수의 원시값에 매핑할 수 있게 된다.
 
 

```r
demo <- tribble(
  ~a,      ~b,
  "bar_1", 20,
  "bar_2", 30,
  "bar_3", 40
)

ggplot(data = demo) +
  geom_bar(mapping = aes(x = a, y = b), stat = "identity")
```

 count가 아니라  proportion으로 보기

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, y = ..prop.., group = 1))
```

```r
 stat_summary() 기능 

각각의 unique x value에 대한 y values를 써머리
```

```r
ggplot(data = diamonds) + 
  stat_summary(
    mapping = aes(x = cut, y = depth),
    fun.ymin = min,
    fun.ymax = max,
    fun.y = median
  )
```

```r
ggplot(data = diamonds, mapping = aes(x = cut, fill = clarity)) + 
  geom_bar(alpha = 1/5, position = "identity")
ggplot(data = diamonds, mapping = aes(x = cut, colour = clarity)) + 
  geom_bar(fill = NA, position = "identity")
```

## 3.9 Coordinate systems

coord_flip()

```r
ggplot(data = mpg, mapping = aes(x = class, y = hwy)) + 
  geom_boxplot()
ggplot(data = mpg, mapping = aes(x = class, y = hwy)) + 
  geom_boxplot() +
  coord_flip()
```

```r
nz <- map_data("nz")

ggplot(nz, aes(long, lat, group = group)) +
  geom_polygon(fill = "white", colour = "black")

ggplot(nz, aes(long, lat, group = group)) +
  geom_polygon(fill = "white", colour = "black") +
  coord_quickmap()
```

coord_polar()

```r
bar <- ggplot(data = diamonds) + 
  geom_bar(
    mapping = aes(x = cut, fill = cut), 
    show.legend = FALSE,
    width = 1
  ) + 
  theme(aspect.ratio = 1) +
  labs(x = NULL, y = NULL)

bar + coord_flip()
bar + coord_polar()
```
